# Scatter Plot 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("penguins")   # Seaborn dataset github 
df
```

```python
sns.scatterplot(x="bill_length_mm",y = "bill_depth_mm",data = df,hue = "sex")
plt.show()   # it is used to remove the heading on the graph 
```

```python
sns.scatterplot(x="bill_length_mm",y = "bill_depth_mm",data = df,hue = "sex",style="sex")
plt.show()   # it is used to remove the heading on the graph 
# style parameter is use to change the style of the parameter
```

```python
sns.scatterplot(x="bill_length_mm",y = "bill_depth_mm",data = df,hue = "sex",style="sex",size ="sex")
# size  = it uses to change the sizes of the male and female 
```

```python
sns.scatterplot(x="bill_length_mm",y = "bill_depth_mm",data = df,hue = "sex",style="sex",size ="sex",sizes = (100,40))
# sizes = it is used to change size according to our convineous 
plt.show()
```
