# What is seaborn 

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.It builds on top of matplotlib and integrates closely with pandas data structures(Data Frame and series).

Seaborn helps you explore and understand your data.
Its plotting functions operate on dataframes and arrays containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots. 
Its dataset-oriented, declarative API lets you focus on what the different elements of your plots mean, rather than on the details of how to draw them.

Seaborn is the only library we need to import for this simple example. By convention, it is imported with the shorthand sns.
Behind the scenes, seaborn uses matplotlib to draw its plots. For interactive work, it’s recommended to use a Jupyter/IPython interface in matplotlib mode, or else you’ll have to call matplotlib.pyplot.show() when you want to see the plot.

2) Pandas

# Line plot in Seaborn 

```python
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,5,6,7,8]

# x_1 = pd.DataFrame({"var":[1,2,3,4,5,6,7],"Var_1":[2,3,4,5,6,7,8]})

x_1 = pd.DataFrame({"var":var,"var_1":var_1})  # we cannot pass data directly we need a dataframe 
sns.lineplot (x="var",y="var_1",data=x_1)        # data parameter is used to pass the data 
plt.show()                                     # <Axes: xlabel='var', ylabel='var_1'> if i want to remove this line then used plt.show   
```

```python
plt.plot(var,var_1)
plt.show()
```

```python
data = pd.read_csv("Iris  (1) (1) (1).csv")
data
```

```python

sns.lineplot (x="SepalLengthCm",y="SepalWidthCm",data=data)        # data parameter is used to pass the data 
plt.show() 
```

```python
sns.lineplot (x="SepalLengthCm",y="SepalWidthCm",data=data,hue ="Species")  
# hue parameter is used to plot the graph with respect some important information 
plt.show() 
```

```python
sns.lineplot (x="SepalLengthCm",y="SepalWidthCm",data=data,hue ="Species",size = 8)  
# size parameter is used to change the size of garph 
plt.show() 
```

```python
sns.lineplot (x="SepalLengthCm",y="SepalWidthCm",data=data,hue ="Species",size = 7,style = "Species")  
# style parameter is used to change the style of the graph  
plt.show() 
```

```python
sns.lineplot (x="SepalLengthCm",y="SepalWidthCm",data=data,hue ="Species",size = 7,style = "Species",palette ='Blues')  
# palette parameter is used to change the Colour of the graph there are different palette for colour 
plt.show() 
```

```python
sns.lineplot (x="SepalLengthCm",y="SepalWidthCm",data=data,hue ="Species",size = 7,style = "Species",palette ="Reds")  
# palette parameter is used to change the Colour of the graph there are different palette for colour Blues , Reds

plt.grid()               # if we want grid the from matplotlib we used grid function 
plt.title("Line Plot")  # it is used to give title which is used from the matplaotlib
plt.show() 
```
