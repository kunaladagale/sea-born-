```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt 
```

to plot the histogram there is one function called as distplot 
in the matplotlib there is histplot 

```python
df= sns.load_dataset("penguins")
df
# we are importing this dataset from the sns directly just writting the load_dataset("dataset name ")
# or u can download it and read it in the jupyter notebbok 
# https://github.com/mwaskom/seaborn-data
# download it and read it in the jupyter notebook 
```

```python
sns.histplot(data=df,x = "bill_length_mm",kde = True)

plt.show()
```

```python
sns.displot(data = df,x = "bill_length_mm")
plt.show()
```

```python
sns.displot(data = df,x = "bill_length_mm",kde =True)
plt.show()
# kde is kearnal density curve
```

```python
sns.displot(data = df,x = "bill_length_mm",kde = True,rug = True,color = "Red")
plt.show()
# rug = rug parameter is used for showing the kernal discity marking 
# Colour parameter is used for change the color
```

```python
sns.displot(data = df,x = "bill_length_mm",kde = True,rug = True,color = "g")
plt.show()
```
